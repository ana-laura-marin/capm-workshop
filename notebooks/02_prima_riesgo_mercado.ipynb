## Prima de Riesgo del Mercado

La prima de riesgo del mercado (Rm − Rf) es el retorno adicional que exige un inversionista 
por invertir en el mercado accionario en lugar de un bono del Tesoro sin riesgo.

**¿Por qué existe esta prima?** Porque las acciones son más riesgosas que los bonos del Tesoro, entonces
el inversionista exige ser compensado por ese riesgo adicional.

Usamos el **S&P 500** como representación del mercado, un índice de las 500 empresas más 
grandes de EE.UU., ampliamente aceptado como proxy del mercado 

**Metodología:** calculamos el retorno compuesto anualizado del S&P 500 para el periodo 
2020–2025 y le restamos el Rf actual.

```python
import sys
sys.path.append('..')

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

from src.db_client import fetch_risk_free_rate
from src.yahoo_client import fetch_returns
```

```python
# Retornos del mercado (S&P 500)
df_mercado = fetch_returns("^GSPC", start="2020-01-01", end="2025-12-31")

# Tasa libre de riesgo
df_rf = fetch_risk_free_rate()

print(f"Datos del S&P 500: {len(df_mercado)} días")
print(df_mercado.head())
```

```python
# Retorno compuesto total del S&P 500 (2020-2025)
retorno_total = (1 + df_mercado['return']).prod() - 1

# Número de años en el periodo
n_años = (df_mercado.index[-1] - df_mercado.index[0]).days / 365.25

# Anualizar el retorno
rm = ((1 + retorno_total) ** (1 / n_años) - 1) * 100

print(f"Retorno total del periodo: {retorno_total*100:.2f}%")
print(f"Años en el periodo: {n_años:.2f}")
print(f"Rm anualizado: {rm:.2f}%")
```

```python
# Rf actual (última observación disponible)
rf = df_rf['rate'].iloc[-1]

print(f"Rf actual (GS10): {rf:.2f}%")

# Prima de riesgo
prima = rm - rf
print(f"Prima de riesgo (Rm - Rf): {prima:.2f}%")
```
